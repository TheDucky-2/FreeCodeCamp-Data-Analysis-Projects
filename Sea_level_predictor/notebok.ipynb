## **Sea Level Predictor**

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
df = pd.read_csv('epa-sea-level.csv')   # loading the csv
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
year = df['Year'] 
sea_level = df['CSIRO Adjusted Sea Level']
```

```python
# plotting a scatter plot 

plt.scatter(year, sea_level)
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.title('Rise in Sea Level')
plt.show()
```

```python
result = linregress(year, sea_level)   # finding the slope and intercept
```

```python
slope = result.slope
intercept = result.intercept

print(slope)
print(intercept)
```

```python
x_val = (list(range(min(year), 2051)))
y_val = [((slope * x) + intercept) for x in x_val]
```

```python
plt.scatter(year, sea_level)   # original scatter plot
plt.plot(x_val, y_val, color = 'red')      # plotting the best line fit
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.title('Rise in Sea Level')
plt.show()
```

```python
# from year 2000

years_from_2000 = year[120:]
sea_level_since_2000 = sea_level[120:]

plt.scatter(years_from_2000, sea_level_since_2000)
plt.show()
```

```python
result_since_2000 = linregress(years_from_2000, sea_level_since_2000)

slope_newline = result_since_2000.slope
intercept_newline = result_since_2000.intercept

x_val_n = (list(range(min(years_from_2000), 2051)))
y_val_n = [((slope_newline * x_n) + intercept_newline) for x_n in x_val_n]
```

```python
plt.scatter(years_from_2000, sea_level_since_2000)
plt.plot(x_val_n, y_val_n, color= 'red', linewidth = 2)
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.title('Rise in Sea Level (2000 - 2050)')
plt.show()
```
